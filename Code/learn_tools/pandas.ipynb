### 数据分组与聚合

```python
sales_df = pd.DataFrame({
    'Product': ['A', 'B', 'A', 'B', 'A'],
    'Region': ['North', 'South', 'North', 'South', 'West'],
    'Sales': [100,150,200,120,180]
})

print(sales_df)
print()
grouped_sales = sales_df.groupby('Product')['Sales'].sum()
# print(grouped_sales)

multi_grouped = sales_df.groupby(['Product', 'Region'])['Sales'].sum()
multi_grouped

agg_result = sales_df.groupby('Product').agg({
    'Sales': ['sum', 'mean', 'count'],
    'Region': 'first'
})

agg_result
```

### 合并数据集

```python
left_df = pd.DataFrame({
    'key': ['K0', 'K1', 'K2'],
    'A': ['A0', 'A1', 'A2']
})

right_df = pd.DataFrame({
    'key': ['K0', 'K1', 'K3'],
    'B': ['B0', 'B1', 'B3']
})

merged_inner = pd.merge(left_df, right_df, on='key', how='inner')
merged_inner

merged_left = pd.merge(left_df, right_df, on='key', how='left')
merged_left
```

### 时间序列

```python
dates = pd.date_range('2023-01-01', periods=10, freq='D')
dates

ts_df = pd.DataFrame({
    'Value': [10,12,11,15,14,16,18,17,19,20]
}, index=dates)
ts_df

weekly_ts = ts_df.resample('W').mean()
weekly_ts

subset = ts_df['2023-01-01': '2023-01-05']
print(subset)
```
