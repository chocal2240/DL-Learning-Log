### 数组操作

```python
arr = np.arange(12)
arr

reshaped_arr = arr.reshape(3,4)
reshaped_arr

reshaped_arr_2 = arr.reshape(2,-1)
reshaped_arr_2
```

```python
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[5,6]])

concatenated = np.concatenate([arr1, arr2], axis=0)
concatenated

arr3 = np.array([[7], [8]])
concatenated_axis1 = np.concatenate([arr1, arr3], axis=1)
concatenated_axis1

v_stack = np.vstack([arr1, arr2])
v_stack
h_stack = np.hstack([arr1, arr3])
h_stack
```

```python
arr = np.arange(12).reshape(3,4)
arr

split_arrays = np.split(arr, 3, axis=0)

split_axis1 = np.split(arr, 2, axis=1)
split_axis1
```

### 数学函数

### 线性代数

```python
import numpy.linalg as la

A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
product = np.dot(A,B)
product

transpose_A = A.T
transpose_A

try:
    inv_A = la.inv(A)
    print(inv_A)
    print("验证：")
    print(np.dot(A, inv_A))
except la.LinAlgError:
    print("No inv")
```
