# What is Pytorch

PyTorch 是一个基于 Python 的科学计算包，服务于两个广泛的目的：
- 作为 NumPy 的替代品，以使用 GPU 和其他加速器的强大功能。
- 一个自动微分库，可用于实现神经网络。

```python
import torch
import numpy as np
```

# **1. Tensors**

张量是一种专门的数据结构，与数组和矩阵非常相似。在 PyTorch 中，我们使用张量来编码模型的输入和输出，以及模型的参数。
在使用上与Python的list、numpy的array、ndarray等数据结构比较类似

## **Tensor Initialization  张量初始化**

.tensor

.from_numpy

.ones_like/rand_like

### 从数据创建(belike list)

```python
data = [[1,2], [3,4]]
x_data = torch.tensor(data)
x_data
```

### 从Numpy数组创建

```python
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
x_np
```

### 从另一个张量创建

```python
x_ones = torch.ones_like(x_data)
print(f"Ones Tensor: \n{x_ones}\n")

x_rand = torch.rand_like(x_data, dtype=torch.float)
print(f"Random Tensor: \n{x_rand}\n")
```

### 随机变量或常量
shape 是一个表示张量维度的元组。在下面的函数中，它决定了输出张量的维数。

.rand

.ones

.zeros


```python
shape = (2, 3,)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor: \n{rand_tensor}\n")
print(f"Ones Tensor: \n{ones_tensor}\n")
print(f"Zeros Tensor: \n{zeros_tensor}\n")
```

## **Tensor Attributes 张量属性**

张量属性描述了它们的形状、数据类型以及存储的设备

.shape

.dtype

.device

```python
tensor = torch.rand(3, 4)

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

## **Tensor Operations 张量操作**

belike 转置、索引、切片、数学运算、线性代数、随机采样

```python
# 转torch运行的设备
if torch.cuda.is_available():
    tensor = tensor.to('cuda')
    print(f"Device tensor is stored on: {tensor.device}")
```

```python
# 标准numpy样式的索引和切片 -- [行选择, 列选择]
tensor = torch.ones(4,4)
tensor[:,1] = 0
print(tensor)
```

```python
# torch.cat连接张量
t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)
```

```python
# 张量的Hadamard积
print(f"tensor.mul(tensor) \n{tensor.mul(tensor)}\n")

print(f"tensor * tensor \n{tensor*tensor}")
```

```python
# 张量的乘法积，以下以行与行之间的点积为例 -- .T即转置
print(f"tensor.matmul(tensor.T)\n{tensor.matmul(tensor.T)}\n")

print(f"tensor @ tensor.T \n{tensor @ tensor.T}")
```

```python
"""
原地操作：以 _ 后缀结尾的操作是原地操作。
例如： x.copy_(y) 、 x.t_() ，会改变 x 。
"""
print(tensor, "\n")
tensor.add_(5)
print(tensor)
```

原地操作可以节省一些内存，但在计算导数时会因立即丢失历史记录而出现问题。因此，不建议使用它们。

## **与Numpy的互换**


### 张量换成Numpy数组

```python
t = torch.ones(5)
print(f"t: {t}")
n = t.numpy()
print(f"n: {n}")
```

### Numpy数组转换为Tensor

```python
n = np.ones(5)
t = torch.from_numpy(n)
```

```python
np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")
```
